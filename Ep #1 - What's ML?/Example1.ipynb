# Imports

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

# Constants definition

```python
EPOCHS = 500
```

# Define and Compile the Neural Network

```python
model = nn.Sequential(nn.Linear(1, 1))

# Keras's default learning rate is 1e-2
optimizer = optim.SGD(model.parameters(), lr=1e-2)

loss_fn = nn.MSELoss()
```

# Providing the data

```python
x = torch.FloatTensor([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0]).reshape((6, 1))
y = torch.FloatTensor([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0]).reshape((6, 1))
```

# Training the Neural Network

```python
for i in range(EPOCHS):
    print(f'Epoch {i}/{EPOCHS}')
    optimizer.zero_grad()
    out = model(x)
    loss = loss_fn(out, y)
    loss.backward()
    optimizer.step()
    print(f'loss: {loss.item():e}')
```

# Evaluating the result

```python
print(model(torch.FloatTensor([[10.0]])))
```
